## Exercise 1

**Build the network of retweets. We will now build a network that has as nodes the Twitter handles of the members of the house, and direct edges between nodes A and B if A has retweeted content posted by B. We will build a weighted network, where the weight of an edge is equal to the number of retweets. You can build the network following the steps below (and you should be able to reuse many of the functions you wrote in previous weeks):**

* **Consider the 200 tweets written by each member of the house (see Data above). For each file, use a regular expression to find retweets and to extract the Twitter handle of the user whose content was retweeted. All retweets begin with "RT @originalAuthor:", where "originalAuthor" is the handle of the user whose content was retweeted (and the part of the text you want to extract).**

* **For each retweet, check if the handle retweeted is the one of a member of the house. If yes, keep it. If no, discard it.**

* **Use a NetworkX DiGraph to store the network. Use weighted edges to account for multiple retweets. Store also the party of each member as a node attribute (use the data in this file, or the data you downloaded in Part 1). Remove self-loops (edges that connect a node with itself).**

```python
url = "https://raw.githubusercontent.com/suneman/socialgraphs2019/master/files/data_twitter/H115_tw_2019.csv"
data1 = pd.read_csv(urllib.request.urlopen(url),sep=",")
```

```python
path = 'C:/Users/lass_/OneDrive/Skrivebord/Socialgrafer/week 8/tweets_2019_2/tweets_2019_full/'

OSname=os.listdir(path)
color_map = []
G = nx.DiGraph()
```

```python
for j in OSname:
    Upath = path + j
    tweetStr=open(Upath,'r',encoding = 'utf-8').read()
    
    k = data1[data1.WikiPageName == j].tw_name.item() #finds the repective twitter name for 'OSname'.
    PartyCol = data1[data1.WikiPageName == j].Party.item() #finds the party.
                     
    #Creates nodes and adds a color.
    if (PartyCol == 'Republican'):
        G.add_node(k, Color = 'blue')
                     
    elif (PartyCol == 'Democratic'):
        G.add_node(k, Color = 'red')
    
    
    
    #Finds all the retweeted names.
    RT=re.findall(r'RT @(\S+)',tweetStr)
   

    for RT_name in RT:
        RT_name=RT_name.replace(':','')
        
        #Cheecks if the Retweeted name is a member of the house
        if RT_name in data1.tw_name.values:
            
            #removes self-loop
            if (k != RT_name):
                
                #adds edge.
                if G.has_edge(RT_name,k):
                    w = G[RT_name][k]['weight']
                    G.add_edge(RT_name, k,weight = w + 1)    
                
                else:
                    G.add_edge(RT_name,k,weight = 1)
```

```python
# Dictorinary of sum of weight of outgoing link for each node:
Weights = {}
for node in G.nodes(data = True):
    edges = G.out_edges(node[0],data = True)
    outgoing = 0
    if edges:
        for edge in edges:
            outgoing = outgoing + edge[2]['weight']
            
    Weights[node[0]] = outgoing

    
# Distribution plot:
sns.kdeplot([Weights[i] for i in dem], label="Democratic")
sns.kdeplot([Weights[i] for i in rep], label="Republican")
plt.legend()
plt.xlabel('Retweets')
plt.xlabel('Probability')
plt.title('Distribution of outgoing strength')
```

## Exercise 4: TF-IDF of the republican and democratic tweets.

**We will create two documents, one containing the words extracted from tweets of republican members, and the other for Democratic members. We will then use TF-IDF to compare the content of these two documents and create a word-cloud. The procedure you should use is exactly the same you used in exercise 2 of week 7. The main steps are summarized below:**

* **Create two large documents, one for the democratic and one for the republican party. Tokenize the pages, and combine the tokens into one long list including all the pages of the members of the same party.**
* **Exclude the twitter handles of other members.**
* **Exclude punctuation.**
* **Exclude stop words (if you don't know what stop words are, go back and read NLPP1e again).**
* **Exclude numbers (since they're difficult to interpret in the word cloud).**
* **Set everything to lower case.**
* **Compute the TF-IDF for each document.**
* **Now, create word-cloud for each party. Are these topics less "boring" than the wikipedia topics? Why? Comment on the results.**

```python
# Get list of members of the house:
url = 'https://raw.githubusercontent.com/suneman/socialgraphs2019/master/files/data_US_congress/H115.csv'
df_WikiPageNames = pd.read_csv(url, usecols = ['WikiPageName','Party'])

# List of democratic members wikipagenames
WikiPageName_demo = list(df_WikiPageNames.groupby('Party').get_group('Democratic').WikiPageName)

# List of republican members wikipagenames
WikiPageName_rep = list(df_WikiPageNames.groupby('Party').get_group('Republican').WikiPageName)

# List of stopwords:
StopWords = corpus.stopwords.words('english')+ ['ahca']+ ['https']

# List of twitter handles:
df_ID = pd.read_csv('../H115_tw_2019.csv')
```

```python
# Two list of sentiment of each tweet for each party:
word_list_demo = []
word_list_rep = []

Tweet_files = glob.glob('../tweets_2019/*')

for i in Tweet_files:
    data_full = open(str(i),'r',encoding = 'utf-8').read()
    
    for data in data_full.split('\n'):   
        # Exclude all hastags, all @.., all &.. and all https links.
        data = re.sub(r'#(\S+)','',data)
        data = re.sub(r'@(\S+)','',data)
        data = re.sub(r'https(\S+)','',data)
        data = re.sub(r'&(\S+)','',data)        
        # Tokenize string:
        string_list = word_tokenize(data)
        # Exclude handles of other members 
        string_list = [b for b in string_list if all(a not in b for a in df_ID.tw_name)]   
        # Exclude punctuations, numbers and set to lower:
        string_list = [word.lower() for word in string_list if word.isalpha()]
        # Exclude stop words:
        string_list = [w for w in string_list if not w in StopWords]
        # Only unique words:
        string_list = np.unique(string_list)

        if i.split('\\')[-1] in WikiPageName_demo:
            word_list_demo.extend(string_list)
        elif i.split('\\')[-1] in WikiPageName_rep:
            word_list_rep.extend(string_list)
```

TD-IDF for both lists with tweets from the democratic party and the republican party is computed below.

```python
# Calculate the TF of the republicans:
TF_rep_words = FreqDist(word_list_rep)
TF_rep = {a:(b/len(word_list_rep)) for (a,b) in TF_rep_words.most_common()}
# Calculate the TF of the democrates:
TF_demo_words = FreqDist(word_list_demo)
TF_demo = {a:(b/len(word_list_demo)) for (a,b) in TF_demo_words.most_common()}

# Calculate the IDF for each word:
IDF = {}
for word in np.unique(word_list_demo + word_list_rep):
    denomi = 0
    if word in word_list_demo: 
        denomi+=1
    if word in word_list_rep: 
        denomi+=1

    IDF[word] = np.log( 2 / (denomi))

    
# Calcualte the TF-IDF of the republicans:
TF_IDF_rep = {}
for word in TF_rep.items():
    TF_IDF_rep[word[0]] = word[1] * IDF[word[0]]
TF_IDF_rep = sorted(TF_IDF_rep.items(), key=lambda kv: kv[1],reverse = True)
# Calcualte the TF-IDF of the democratics:
TF_IDF_demo = {}
for word in TF_demo.items():
    TF_IDF_demo[word[0]] = word[1] * IDF[word[0]]
TF_IDF_demo = sorted(TF_IDF_demo.items(), key=lambda kv: kv[1],reverse = True)
```

Two wordclouds for each party in generated below. The hold the most frequent word based on the TF-IDF calculations. I do not think that the topics/words in the wordclouds are more boring than the one generated based on the wikipedia pages. The wordclouds of the wikipedia pages holds words like months, names, states and towns - which are find more boring.

```python
mask = np.array(Image.open('./usa_mask.jpg'))

plt.figure(figsize = [25,25])
plt.subplot(1,2,1)
plt.imshow(WordCloud(stopwords=set(list(STOPWORDS) + StopWords),
                     background_color='white',
                     max_words = 50, colormap = 'Blues_r',
                     mask=mask)
           .generate_from_frequencies(dict(TF_IDF_demo)));
plt.axis('off');
plt.title('Democratic',fontsize = 36);

plt.subplot(1,2,2)
plt.imshow(WordCloud(stopwords=set(list(STOPWORDS) + StopWords),
                     background_color='white',
                     max_words = 50, colormap = 'Reds_r',
                     mask=mask)
           .generate_from_frequencies(dict(TF_IDF_rep)));
plt.axis('off');
plt.title('Republican',fontsize = 36);
```

### Sentiment wordlist and function:

The data seen below hold a number of measures for 10,222 different words. It is generated using Crowdsourcing from Amazon Mechanical Turk, which includes that real human have rated the words according to happiness. The happiness attributes will be the one that will be used as a measure of sentiment in the following exercises.

```python
df = pd.read_csv('Data_set_S1.csv',sep='\t')
df.head(5)
```

The following implements a function that calculates sentiment given a list of tokens.

```python
# Sentiment dictonary:
sentiments = dict(zip(df.word,df.happiness_average))
```

```python
# Function that calculates sentiment given a list of tokens
def sentiment_func(token_list):
    res = []
    for word in token_list:
        if word in df.word.values:
            res.append(sentiments[word.lower()])
        
    return sum(res)/len(res) if len(res) else 0
```

### Preprocession of tweets:
The following process the tweets by generating two list of tweets: one including tweets by democratic members, and the other including tweets of republican members. The sentiment of each tweet is also calculated and stored in seperate lists.

```python
# Get the list twitter accounts:
df_ID = pd.read_csv('../H115_tw_2019.csv')
# Dictonary to translate reale name to Twitter account
Names2Twitter = dict(df_ID[['WikiPageName','tw_name']].to_numpy())
# Dictonary to translate Twitter acount to real name
Twitter2Names = dict(df_ID[['tw_name','WikiPageName']].to_numpy())
# List of tweet files
Tweet_files = glob.glob('../tweets_2019/*')
```

```python
# Get list of members of the house:
url = 'https://raw.githubusercontent.com/suneman/socialgraphs2019/master/files/data_US_congress/H115.csv'
df_WikiPageNames = pd.read_csv(url, usecols = ['WikiPageName','Party'])
# List of democratic members wikipagenames
WikiPageName_demo = list(df_WikiPageNames.groupby('Party').get_group('Democratic').WikiPageName)
# List of republican members wikipagenames
WikiPageName_rep = list(df_WikiPageNames.groupby('Party').get_group('Republican').WikiPageName)
```

```python
# List of stopwords:
StopWords = corpus.stopwords.words('english') + ['ahca'] + ['https']
```

```python
# Two list of sentiment of each tweet for each party:
sent_list_demo = []
sent_list_rep = []
word_list_demo = []
word_list_rep = []

for i in Tweet_files:
    data_full = open(str(i),'r',encoding = 'utf-8').read()
    
    for data in data_full.split('\n'):
        # Exclude all hastags, all @.., all &.. and all https links.
        data = re.sub(r'#(\S+)','',data)
        data = re.sub(r'@(\S+)','',data)
        data = re.sub(r'https(\S+)','',data)
        data = re.sub(r'&(\S+)','',data) 
        # Tokenize string:
        string_list = word_tokenize(data)
        # Exclude handles of other members 
        string_list = [b for b in string_list if all(a not in b for a in df_ID.tw_name)]        
        # Exclude punctuations, numbers and set to lower:
        string_list = [word.lower() for word in string_list if word.isalpha()]
        # Exclude stop words:
        string_list = [w for w in string_list if not w in StopWords]
        
        # Check is list is empty:
        if string_list:
            if i.split('\\')[-1] in WikiPageName_demo:
                sent_list_demo.append(sentiment_func(string_list))
                word_list_demo.append(string_list)
            elif i.split('\\')[-1] in WikiPageName_rep:
                sent_list_rep.append(sentiment_func(string_list))
                word_list_rep.append(string_list)
```

### Generel Tweet statistics:
The following shows the distribution of the sentiment of the democratic and republican tweets. The distribution of the two parties are very similar - they both post most positive tweets. However the republican party seems to be a bit more postive than the democratic party.

```python
# Distribution plot:
sns.kdeplot(sent_list_demo, label="Democratic")
sns.kdeplot(sent_list_rep, label="Republican")
plt.legend();
plt.xlabel('Sentiment');
plt.ylabel('Probability');
plt.title('Distribution of sentiment');

# The average m and standard deviation σ of the Tweets sentiment (considering tweets by both republican and democrats)
m = np.mean(sent_list_demo + sent_list_rep)
print('The mean sentiment is: ' + str(m))

sigma = np.std(sent_list_demo + sent_list_rep)
print('The standard deviation if the sentiment is: ' + str(sigma))
```

### Negative Tweets:

In the following only negative tweets from the two parties will be analysed. The analysis is result of the tweets with a sentiment lower than m−2σ. For each party a wordcloud is generated based on the negative tweets. TF-IDF is computed for this purpose. It is seen that the word clouds don't have many words in common and that the words are negative in my oppinion.

```python
# List of negative tweets by the two Parties:
negative_demo = [word_list_demo[i] for i in range(len(word_list_demo)) if sent_list_demo[i] < m-(2*sigma)]
negative_rep = [word_list_rep[i] for i in range(len(word_list_rep)) if sent_list_rep[i] < m-(2*sigma)]
```

```python
#Calculate the TF of the republicans:
TF_rep_words = FreqDist(np.concatenate(negative_rep))
TF_rep = {a:(b/len(np.concatenate(negative_rep))) for (a,b) in TF_rep_words.most_common()}
#Calculate the TF of the democratics:
TF_demo_words = FreqDist(np.concatenate(negative_demo))
TF_demo = {a:(b/len(np.concatenate(negative_demo))) for (a,b) in TF_demo_words.most_common()}

# Calculate the IDF for each word:
IDF = {}
for word in np.unique(np.concatenate([np.concatenate(negative_demo), np.concatenate(negative_rep)])):
    denomi = 0
    if word in np.concatenate(negative_rep): 
        denomi+=1
    if word in np.concatenate(negative_demo): 
        denomi+=1

    IDF[word] = np.log( 2 / (denomi))
```

```python
# Calcualte the TF-IDF of the republicans:
TF_IDF_rep = {}
for word in TF_rep.items():
    TF_IDF_rep[word[0]] = word[1] * IDF[word[0]]
TF_IDF_rep = sorted(TF_IDF_rep.items(), key=lambda kv: kv[1],reverse = True)
```

```python
# Calcualte the TF-IDF of the democratics:
TF_IDF_demo = {}
for word in TF_demo.items():
    TF_IDF_demo[word[0]] = word[1] * IDF[word[0]]
TF_IDF_demo = sorted(TF_IDF_demo.items(), key=lambda kv: kv[1],reverse = True)
```

Below are the two word clouds of the negative tweets made by the two Parties:

```python
mask = np.array(Image.open('./twitter_mask.png'))
plt.figure(figsize = [25,25])
plt.subplot(1,2,1)
plt.imshow(WordCloud(stopwords=set(list(STOPWORDS) + StopWords),
                     background_color='white',
                     max_words = 50, colormap = 'Blues_r',mask = mask)
           .generate_from_frequencies(dict(TF_IDF_demo)));
plt.axis('off')
plt.title('Democratic',fontsize = 36);

plt.subplot(1,2,2)
plt.imshow(WordCloud(stopwords=set(list(STOPWORDS) + StopWords),
                     background_color='white',
                     max_words = 50, colormap = 'Reds_r',mask = mask)
           .generate_from_frequencies(dict(TF_IDF_rep)));
plt.axis('off');
plt.title('Republican',fontsize = 36);
```
